```python
import psycopg2
import pandas as pd
import os # Import the os module for working with file paths
```

```python
def create_database():
    # connect to default database
    conn = psycopg2.connect(
    host="postgres_pgadmin_postgres_1", #Postgres Container Names
    database="postgres",
    user="postgres",
    password="postgres")
    cur = conn.cursor()
    conn.set_session(autocommit=True)
    
    # Forcibly Terminate Connections
    cur.execute("SELECT pg_terminate_backend(pg_stat_activity.pid) \
    FROM pg_stat_activity \
    WHERE pg_stat_activity.datname = 'accounts';")
    
    # create sparkify database with UTF8 encoding
    cur.execute("SELECT 1 FROM pg_database WHERE datname = 'accounts'")
    
    if cur.fetchone():
        cur.execute("DROP DATABASE accounts")
    
    #cur.execute("DROP DATABASE accounts")
    cur.execute("CREATE DATABASE accounts")
    
    # close connection to default database
    conn.close()
    
    # connect to sparkify database
    conn = psycopg2.connect(
    host="postgres_pgadmin_postgres_1", #Postgres Container Names
    database="accounts",
    user="postgres",
    password="postgres")
    cur = conn.cursor()
    
    return cur, conn
```
